## Identifying Columns with Uniform Values Across Car Listings

```python
# Identifying columns with only one unique value across all listings, which may not provide meaningful insights

all_columns = df.columns
unusefull_column_one_unique_value =[]
for column in all_columns:
    try :
        values = df[column].unique()
        num_values = len(values)
    except :
        values = df[column].explode()
        num_values = len(values.unique())
    if num_values == 1 :
        unusefull_column_one_unique_value.append(column)
```

## Detecting Columns with a Significant Number of Missing Values Across Car Listings

```python
all_columns = df.columns
unusefull_column_many_missing_values ={}
for column in all_columns:
    try :
        values = df[column].count()
    except :
        values = df[column].explode()
        values = values.count()
    if values < 1700 :
        unusefull_column_many_missing_values[column]= values
```

## List of Columns with Substantial Missing Values Across Car Listings

```python
unusefull_column_many_missing_values = list(unusefull_column_many_missing_values.keys())
```

## Removing Columns with Significant Missing Values from the DataFrame

```python
df = df.drop(columns= unusefull_column_many_missing_values, axis = 1)
```

## Removing Columns with Uniform Values Across Car Listings from the DataFrame

```python
df = df.drop(columns= unusefull_column_one_unique_value, axis = 1)
```
